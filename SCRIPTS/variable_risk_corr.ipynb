```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('sdoh.csv', encoding='windows-1254')
```

```python
df1
```

```python
from sentence_transformers import SentenceTransformer, util

model2 = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
variables = df1['Variable Label'].to_list()
```

```python
embedding1 = model2.encode(variables, convert_to_tensor=False)
```

```python
embedding1.shape
```

```python
np.save('emb_var_label', embedding1) 
```

```python
emb = pd.DataFrame(embedding1)
emb['Variable Name'] = df1['Variable Name']
```

```python
emb
```

```python
emb.to_json('emb.json')
```

```python
risks = ['Transportation Risk: Features related to commuting/nearest available hospital, such as commute distance, availability of public/private transport, number of available hospitals in locality, that may impact access to healthcare facilities and services',
         'Technology Access Risk: The risk associated with limited access to technology, including the internet and cell phones, which may hinder the ability to receive timely health-related information',
         'Financial Risk: The potential health impact arising from factors such as lack of medical insurance, low income, or unemployment',
         'Food Security: The risk associated with insufficient availability of food in the neighbourhood or a lack of stores, affecting access to nutritious meals',
         'Housing Challenges: Risks related to housing conditions, including lack of housing, presence of mold, and water leaks, which may impact overall health and well-being',
         'Disease Risk: Identification of prevalent diseases based on location and demographics, contributing to the understanding of potential health threats in a specific population',
         'Genetic Risk: Health risks associated with an individual\'s genetic predispositions and past health history',
         'Climate Risk: Risks associated with weather conditions, such as heat strokes in high-temperature areas, affecting health and well-being',
         'Education Challenges: Risks related to the lack of access to schools and education, impacting overall health and socio-economic well-being',
         'Social Environment Risk: Risks arising from the social environment, including challenges from family members, drug/alcohol abuse, and mental health issues',
         'Lifestyle Risk: Risks related to lifestyle factors, such as sleep duration and physical activity, that can impact overall health and contribute to various health conditions',
         'Health Accessibility Risk: The health risk associated with a lack of medical facilities such as hospitals, clinics']
```

```python
embedding2 = model2.encode(risks, convert_to_tensor=False)
```

```python
emb2 = pd.DataFrame(embedding2)
emb2['Risk'] = ['Transportation Risk', 'Technology Access Risk', 
                      'Financial Risk','Food Security',
                      'Housing Challenges','Disease Risk',
                      'Genetic Risk','Climate Risk',
                      'Education Challenges','Social Environment Risk',
                      'Lifestyle Risk', 'Healthcare Accessibility Risk']
```

```python
emb2
```

```python
emb2.to_json('emb2.json')
```

```python
cosine_scores = util.cos_sim(embedding1, embedding2)
```

```python
df2 = pd.DataFrame(cosine_scores)
```

```python
df2
```

```python
df2.rename(columns = {0:'Transportation Risk', 1:'Technology Access Risk', 
                      2:'Financial Risk',3:'Food Security',
                      4:'Housing Challenges',5:'Disease Risk',
                      6:'Genetic Risk',7:'Climate Risk',
                      8:'Education Challenges',9:'Social Environment Risk',
                      10:'Lifestyle Risk', 11:'Healthcare Accessibility Risk'}, inplace = True)
```

```python
df2
```

```python
df3 = df1[['Variable Name','Variable Label','County 2020', 'ZIP Code 2020', 'Census Tract 2020']]
df3
```

```python
df4 = pd.merge(df3, df2, left_index=True, right_index=True)
df4
```

```python
di = {'X':1, np.nan:0}
df4.replace({'County 2020': di, 'ZIP Code 2020':di, 'Census Tract 2020':di}, inplace=True)
df4
```

```python
df4.nlargest(40, ['Healthcare Accessibility Risk'])[['Variable Label','Healthcare Accessibility Risk']]
```
